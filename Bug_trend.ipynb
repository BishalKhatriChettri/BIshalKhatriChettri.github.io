```python

# Bug Trend Analysis - QA Report
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For better plot style
sns.set(style="whitegrid")

# Read CSV directly from GitHub (RAW link)
url = "https://raw.githubusercontent.com/BishalKhatriChettri/BIshalKhatriChettri.github.io/main/bug_reports.csv"
df = pd.read_csv(url)

# Convert date columns
df['ReportedDate'] = pd.to_datetime(df['ReportedDate'], errors='coerce')
df['ClosedDate'] = pd.to_datetime(df['ClosedDate'], errors='coerce')

# Show data
print("ðŸ“Š First few bug records:")
print(df.head())

# Plot 1: Bug count by severity
plt.figure(figsize=(6,4))
df['Severity'].value_counts().plot(kind='bar', color='orange')
plt.title('Bug Count by Severity')
plt.xlabel('Severity')
plt.ylabel('Number of Bugs')
plt.tight_layout()
plt.show()

# Plot 2: Bugs reported over time
plt.figure(figsize=(8,4))
df.groupby('ReportedDate').size().plot(marker='o')
plt.title('Bugs Reported Over Time')
plt.xlabel('Reported Date')
plt.ylabel('Bug Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 3: Bugs by module
plt.figure(figsize=(6,4))
df['Module'].value_counts().plot(kind='barh', color='green')
plt.title('Bug Count by Module')
plt.xlabel('Number of Bugs')
plt.tight_layout()
plt.show()

# Plot 4: Bugs per tester
plt.figure(figsize=(6,4))
sns.barplot(x=df['Tester'].value_counts().index, y=df['Tester'].value_counts().values)
plt.title('Bugs Found by Each Tester')
plt.xlabel('Tester')
plt.ylabel('Bug Count')
plt.tight_layout()
plt.show()
```
